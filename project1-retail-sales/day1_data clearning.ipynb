```python
# Convert Order Date from text to actual date format
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Convert Ship Date from text to actual date format
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Verify the change worked
print(df[['Order Date', 'Ship Date']].dtypes)
```

```python
# Extract Year from Order Date
df['Order Year'] = df['Order Date'].dt.year

# Extract Month Number (1=January, 12=December)
df['Order Month'] = df['Order Date'].dt.month

# Extract Month Name (January, February...)
df['Order Month Name'] = df['Order Date'].dt.month_name()

# Extract Quarter (Q1, Q2, Q3, Q4)
df['Order Quarter'] = df['Order Date'].dt.quarter

# Calculate how many days it took to ship the order
df['Days to Ship'] = (df['Ship Date'] - df['Order Date']).dt.days

# Verify these new columns look correct
df[['Order Date', 'Ship Date', 'Order Year', 'Order Month', 'Order Month Name', 'Order Quarter', 'Days to Ship']].head(10)
```

```python
# Strip extra whitespace from all text columns
text_columns = df.select_dtypes(include='object').columns

for col in text_columns:
    df[col] = df[col].str.strip()

print("Whitespace cleaned from all text columns.")
```

```python
# Standardize text casing for important columns
df['Segment'] = df['Segment'].str.title()     # Consumer, Corporate
df['Region'] = df['Region'].str.title()       # East, West, North, South
df['Category'] = df['Category'].str.title()   # Furniture, Technology
df['Ship Mode'] = df['Ship Mode'].str.title() # Standard Class, etc.
```

```python
# Profit Margin % â€” shows how profitable each order is
df['Profit Margin %'] = round((df['Profit'] / df['Sales']) * 100, 2)

# Revenue after Discount â€” actual revenue received
df['Revenue After Discount'] = df['Sales'] * (1 - df['Discount'])

# Is the order profitable? (True/False flag â€” easy for filtering)
df['Is Profitable'] = df['Profit'] > 0

# Discount Band â€” categorize discounts into buckets
def discount_band(discount):
    if discount == 0:
        return 'No Discount'
    elif discount <= 0.20:
        return 'Low (1-20%)'
    elif discount <= 0.40:
        return 'Medium (21-40%)'
    else:
        return 'High (41%+)'

df['Discount Band'] = df['Discount'].apply(discount_band)

# Shipping Speed Category
def ship_speed(days):
    if days <= 2:
        return 'Fast (1-2 days)'
    elif days <= 4:
        return 'Normal (3-4 days)'
    else:
        return 'Slow (5+ days)'

df['Shipping Speed'] = df['Days to Ship'].apply(ship_speed)

# Verify all new columns
df[['Sales', 'Profit', 'Profit Margin %', 'Revenue After Discount', 
    'Is Profitable', 'Discount Band', 'Days to Ship', 'Shipping Speed']].head(10)
```

```python
print("=" * 50)
print("DATA QUALITY REPORT")
print("=" * 50)

print(f"\nâœ… Total Rows: {len(df)}")
print(f"âœ… Total Columns: {len(df.columns)}")
print(f"âœ… Missing Values Remaining: {df.isnull().sum().sum()}")
print(f"âœ… Duplicate Rows Remaining: {df.duplicated().sum()}")
print(f"âœ… Date Range: {df['Order Date'].min().date()} to {df['Order Date'].max().date()}")
print(f"âœ… Years Covered: {sorted(df['Order Year'].unique())}")
print(f"âœ… Total Sales Value: ${df['Sales'].sum():,.2f}")
print(f"âœ… Total Profit: ${df['Profit'].sum():,.2f}")
print(f"âœ… Profitable Orders: {df['Is Profitable'].sum()} out of {len(df)}")
print(f"âœ… Loss-Making Orders: {(~df['Is Profitable']).sum()}")
print(f"âœ… Average Days to Ship: {df['Days to Ship'].mean():.1f} days")
print(f"\nðŸ“¦ Orders by Segment:")
print(df['Segment'].value_counts())
print(f"\nðŸ—ºï¸ Orders by Region:")
print(df['Region'].value_counts())
print(f"\nðŸ“ Orders by Category:")
print(df['Category'].value_counts())
print(f"\nðŸ’¸ Orders by Discount Band:")
print(df['Discount Band'].value_counts())
```

```python
# Are there orders where Sales is 0 or negative? 
suspicious_sales = df[df['Sales'] <= 0]
print(f"Suspicious Sales rows: {len(suspicious_sales)}")

# Are there orders with Profit Margin below -100%? (Extreme loss)
extreme_loss = df[df['Profit Margin %'] < -100]
print(f"Extreme loss orders: {len(extreme_loss)}")
if len(extreme_loss) > 0:
    print(extreme_loss[['Order ID', 'Sales', 'Profit', 'Profit Margin %', 'Discount']])

# Are there any shipping times that are negative? (Ship before order â€” impossible)
negative_shipping = df[df['Days to Ship'] < 0]
print(f"Negative shipping days: {len(negative_shipping)}")
```

```python
# Flag suspicious rows instead of deleting them (always preserve data for client)
df['Data Quality Flag'] = 'Clean'
df.loc[df['Profit Margin %'] < -100, 'Data Quality Flag'] = 'Review - Extreme Loss'
df.loc[df['Days to Ship'] < 0, 'Data Quality Flag'] = 'Review - Shipping Date Error'

print(df['Data Quality Flag'].value_counts())
```

```python
# Save as CSV (universal format, works everywhere)
df.to_csv('superstore_cleaned.csv', index=False)
print("âœ… Cleaned CSV saved!")

# Save as Excel (clients prefer this format)
df.to_excel('superstore_cleaned.xlsx', index=False)
print("âœ… Cleaned Excel file saved!")

# Verify the saved file loaded back correctly
verify = pd.read_csv('superstore_cleaned.csv')
print(f"âœ… Verification: {len(verify)} rows, {len(verify.columns)} columns loaded successfully")
```
